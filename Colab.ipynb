```python
from TTS.api import TTS
import torch

print("CUDA Available:", torch.cuda.is_available())

# Get the model manager
tts_manager = TTS().list_models()

# Now call .list_models() again to get the actual list
all_models = tts_manager.list_models()

print("\nAvailable TTS Models:\n")
for idx, model in enumerate(all_models, 1):
    print(f"{idx}. {model}")
```

```python
import ipywidgets as widgets
from IPython.display import display
from TTS.api import TTS

model_manager = TTS().list_models()
model_list = model_manager.list_models()

dropdown = widgets.Dropdown(
    options=model_list,
    value='tts_models/multilingual/multi-dataset/xtts_v2',
    description='Choose model:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='55%')
)

display(dropdown)
```
